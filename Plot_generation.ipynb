```python
from google.colab import drive

drive.mount('/content/gdrive', force_remount = True)
dataset_path = 'gdrive/My Drive/'
```

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_csv(os.path.join(dataset_path, 'Table-2020_43.csv'))
df1.head()
```

```python
x = df1['2013-14'].tolist()
y = df1['2014-15'].tolist()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c='red', label='Import export correlation',
               alpha=0.3, edgecolors='none')
ax.legend()
ax.set_xlabel('Quantity in 2013-14')
ax.set_ylabel('Quantity in 2014-15')
ax.grid(True)
ax.set_title('Correlation between import/export in 2013-14 and 2014-15')
plt.show()
fig.savefig(os.path.join(dataset_path, 'scatter.png'))
```

```python
df2 = pd.read_csv(os.path.join(dataset_path, 'RS_Session_252_AU_236.csv'))
df2.head()
```

```python
states = df2['State-wise'].tolist()
doctors = df2['Number of Registered Doctors'].tolist()
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)
ax.barh(states, doctors)
ax.set_ylabel('States')
ax.set_xlabel('Number of Registered Doctors')
for i, v in enumerate(doctors):
    ax.text(v + 3, i, str(v), color='blue', fontweight='bold')
ax.grid(True)
ax.set_title('Analysis of number of registered doctors in each state')
plt.show()
fig.savefig(os.path.join(dataset_path, 'bar.png'))
```
